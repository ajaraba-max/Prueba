```python
def identificar_origen(nombre):
  """
  Identifica el origen de un personaje de AstÃ©rix segÃºn la terminaciÃ³n de su nombre.
  """
  # Convertimos el nombre a minÃºsculas para una comparaciÃ³n sin errores.
  nombre_lower = nombre.lower()

  if nombre_lower.endswith('ix'):
    return "Â«GaloÂ»"
  elif nombre_lower.endswith('us'):
    return "Â«RomanoÂ»"
  elif nombre_lower.endswith('ic'):
    return "Â«GodoÂ»"
  elif nombre_lower.endswith('as'):
    return "Â«GriegoÂ»"
  elif nombre_lower.endswith('af'):
    return "Â«NormandoÂ»"
  elif nombre_lower.endswith('is') or nombre_lower.endswith('ax'):
    return "Â«BretonÂ»"
  elif nombre_lower.endswith('ez'):
    return "Â«HispanoÂ»"
  # Esta es la Ãºltima comprobaciÃ³n de sufijo para evitar conflictos con "as".
  elif nombre_lower.endswith('a'):
    return "Â«IndioÂ»"
  else:
    return "Â«DesconocidoÂ»"

def main():
  """
  FunciÃ³n principal para leer los datos de entrada y mostrar los resultados.
  """
  try:
    # Leemos el nÃºmero de nombres que se van a introducir.
    n_nombres = int(input())

    # Creamos una lista para guardar los nombres.
    nombres = []
    for _ in range(n_nombres):
      nombres.append(input())

    # Procesamos cada nombre y mostramos su origen.
    for nombre in nombres:
      origen = identificar_origen(nombre)
      print(origen)

  except (ValueError, EOFError):
    # Manejo de errores si la entrada no es un nÃºmero o estÃ¡ vacÃ­a.
    print("Entrada invÃ¡lida. Por favor, ingrese un nÃºmero seguido de los nombres.")

# Ejecutamos el programa.
if __name__ == "__main__":
  main()
```

```python
import random

print("Bienvenido a Conecta4\n")

# Funcion para mostrar el tablero:
def imprimir_tablero(tablero):
  print(" ", end="")
  for i in range(columnas):
    print(f"{i+1} ", end="")

  print()
  print("+", end="")
  for i in range(columnas-1):
    print(f"- ", end="")
  print("-+", end="")

  print()
  for i in tablero:
    print("|", end="")
    print(*i, end="")
    print("|", end="")
    print()

  print("+", end="")
  for i in range(columnas-1):
    print(f"- ", end="")
  print("-+")

# FunciÃ³n para devolver el tablero como cadena para guardarlo en archivo
def tablero_a_string(tablero):
  salida = ""
  salida += " "
  for i in range(len(tablero[0])):
    salida += f"{i+1} "
  salida += "\n+"
  for i in range(len(tablero[0])-1):
    salida += "- "
  salida += "-+\n"
  for fila in tablero:
    salida += "|" + "".join(fila) + "|\n"
  salida += "+"
  for i in range(len(tablero[0])-1):
    salida += "- "
  salida += "-+\n"
  return salida

# Verificar condiciones de victoria
def verificar_victoria(tablero, ficha):
  for fila in range(filas):
    for col in range(columnas - 3):
      if tablero[fila][col] == ficha and tablero[fila][col+1] == ficha and tablero[fila][col+2] == ficha and tablero[fila][col+3] == ficha:
        return True

  for fila in range(filas - 3):
    for col in range(columnas):
      if tablero[fila][col] == ficha and tablero[fila+1][col] == ficha and tablero[fila+2][col] == ficha and tablero[fila+3][col] == ficha:
        return True

  for fila in range(filas - 3):
    for col in range(columnas - 3):
      if tablero[fila][col] == ficha and tablero[fila+1][col+1] == ficha and tablero[fila+2][col+2] == ficha and tablero[fila+3][col+3] == ficha:
        return True

  for fila in range(3, filas):
    for col in range(columnas - 3):
      if tablero[fila][col] == ficha and tablero[fila-1][col+1] == ficha and tablero[fila-2][col+2] == ficha and tablero[fila-3][col+3] == ficha:
        return True

  return False

def tablero_lleno(tablero):
  for fila in tablero:
    if "." in fila:
      return False
  return True

def validacion_jugadores(jugador_num):
  while True:
    nombre = input(f"Por favor, ingrese el nombre del jugador {jugador_num}: ")
    if not nombre:
      print("El nombre no puede estar vacÃ­o. IntÃ©ntelo de nuevo.")
    elif len(nombre) > 10:
      print("El nombre no puede exceder los 10 caracteres. IntÃ©ntelo de nuevo.")
    elif not nombre.isalpha():
      print("El nombre solo puede contener letras. IntÃ©ntelo de nuevo.")
    else:
      return nombre

def solicitar_dimensiones():
  while True:
    try:
      f = int(input("Ingresa el nÃºmero de filas: "))
      c = int(input("Ingresa el nÃºmero de columnas: "))
      print()
      if f <= 3 or c <= 3:
        print("El nÃºmero de filas y/o columnas no es vÃ¡lido.\n")
      else:
        return f, c
    except ValueError:
      print("Entrada invÃ¡lida. Por favor, ingresa un nÃºmero entero.\n")

# Solicitar nombres
print("Nombre del primer jugador:")
nombre_jugador1 = validacion_jugadores(1)
print("Nombre del segundo jugador:")
nombre_jugador2 = validacion_jugadores(2)
print()

# EstadÃ­sticas
partidas_jugadas = 0
victorias_j1 = 0
victorias_j2 = 0
empates = 0
historial_partidas = []

# Bucle de rondas
while True:
  if random.choice([True, False]):
    jugador_actual = nombre_jugador1
    ficha_actual = "X"
    siguiente_jugador = nombre_jugador2
    siguiente_ficha = "O"
  else:
    jugador_actual = nombre_jugador2
    ficha_actual = "O"
    siguiente_jugador = nombre_jugador1
    siguiente_ficha = "X"

  print(f"{jugador_actual} ha sido elegido para comenzar.\n")

  filas, columnas = solicitar_dimensiones()
  Tablero = [["."] * columnas for _ in range(filas)]
  imprimir_tablero(Tablero)

  resultado_partida = ""
  while True:
    jugador = jugador_actual
    ficha = ficha_actual

    print()
    while True:
      modo = input(f"{jugador}, Â¿deseas jugar en modo 'manual' o 'aleatorio'? ").lower()
      if modo not in ["manual", "aleatorio"]:
        print("Modo invÃ¡lido. Escribe 'manual' o 'aleatorio'.")
      else:
        break

    if modo == "manual":
      jugada = input(f"{jugador}, elige una columna (1 a {columnas}): ")
      if jugada == "":
        print("Debes ingresar un nÃºmero de columna.")
        continue
      if not jugada.isdigit():
        print("Solo se permiten nÃºmeros enteros.")
        continue
      col = int(jugada) - 1
    else:
      print(f"{jugador} (modo aleatorio) estÃ¡ eligiendo una columna...")
      posibles_columnas = [i for i in range(columnas) if Tablero[0][i] == "."]
      if not posibles_columnas:
        print("No hay mÃ¡s movimientos posibles. Â¡Empate!")
        break
      col = random.choice(posibles_columnas)
      print(f"{jugador} eligiÃ³ la columna {col+1}")

    if col < 0 or col >= columnas:
      print("Columna fuera de rango. Intenta de nuevo.")
      continue

    fila_disponible = -1
    for i in range(filas - 1, -1, -1):
      if Tablero[i][col] == ".":
        fila_disponible = i
        break

    if fila_disponible == -1:
      print("Esa columna estÃ¡ llena. Elige otra.")
      continue

    Tablero[fila_disponible][col] = ficha
    imprimir_tablero(Tablero)

    if verificar_victoria(Tablero, ficha):
      print(f"\nðŸŽ‰ Â¡{jugador} ha ganado el juego con la ficha '{ficha}'! ðŸŽ‰\n")
      partidas_jugadas += 1
      if jugador == nombre_jugador1:
        victorias_j1 += 1
      else:
        victorias_j2 += 1
      resultado_partida = f"{jugador} ganÃ³ con la ficha '{ficha}'"
      break

    if tablero_lleno(Tablero):
      print("\nÂ¡El tablero estÃ¡ lleno! Se declara un empate.\n")
      partidas_jugadas += 1
      empates += 1
      resultado_partida = "Empate"
      break

    jugador_actual, siguiente_jugador = siguiente_jugador, jugador_actual
    ficha_actual, siguiente_ficha = siguiente_ficha, ficha_actual

  # Guardar tablero y resultado en historial
  historial_partidas.append({
    "tablero": [fila[:] for fila in Tablero],
    "resultado": resultado_partida
  })

  # Â¿Jugar otra ronda?
  while True:
    continuar = input("Â¿Desean jugar otra ronda? (si/no): ").strip().lower()
    if continuar == "si":
      print("\nIniciando una nueva partida...\n")
      break
    elif continuar == "no":
      print("\nGracias por jugar Conecta4. Generando estadÃ­sticas...\n")

      with open("estadisticas_conecta4.txt", "w", encoding="utf-8") as archivo:
        archivo.write("=== EstadÃ­sticas de Conecta4 ===\n\n")
        archivo.write(f"Jugador 1: {nombre_jugador1} (X)\n")
        archivo.write(f"Jugador 2: {nombre_jugador2} (O)\n\n")
        archivo.write(f"Partidas jugadas: {partidas_jugadas}\n")
        archivo.write(f"Victorias de {nombre_jugador1}: {victorias_j1}\n")
        archivo.write(f"Victorias de {nombre_jugador2}: {victorias_j2}\n")
        archivo.write(f"Empates: {empates}\n\n")

        for idx, partida in enumerate(historial_partidas, 1):
          archivo.write(f"--- Partida {idx} ---\n")
          archivo.write(tablero_a_string(partida["tablero"]))
          archivo.write(f"Resultado: {partida['resultado']}\n\n")

      print("ðŸ“„ Archivo 'estadisticas_conecta4.txt' generado con Ã©xito.")
      print("Â¡Hasta la prÃ³xima! ðŸ‘‹")
      exit()
    else:
      print("Respuesta invÃ¡lida. Escribe 'si' o 'no'.")
```
